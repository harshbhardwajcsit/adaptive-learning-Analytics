```python
#Instructor description
import pandas as pd
import seaborn as sns  
import matplotlib
import matplotlib.pyplot as plt  
matplotlib.style.use('ggplot')
%matplotlib inline
#number of courses by institution
data = pd.read_csv('appendix.csv')


# fig specification
fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.3, wspace=0.2)
fig.suptitle('no. of the courses launched by each Intitute on give date ', fontsize=15, fontweight='bold')

no_of_participents = data[['Institution',"Participants (Course Content Accessed)",'Launch Date']].groupby(['Institution','Launch Date']).sum()
no_of_participents = no_of_participents.reset_index()

print(no_of_participents)
sns.barplot(x='Launch Date',y='Participants (Course Content Accessed)',hue='Institution',data=no_of_participents)
```
