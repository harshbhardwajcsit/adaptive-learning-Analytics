```python
#visaulization of count with single category variable


import pandas as pd
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('xAPI-Edu-Data.csv')

# defining the category variables
cat_Vars = ['gender', 'NationalITy', 'PlaceofBirth', 'StageID','Relation', 'GradeID', 'Topic', 'Semester', 'StudentAbsenceDays']

# fig specification
fig = plt.figure(figsize=(12, 5))
fig.subplots_adjust(hspace=.3, wspace=0.2)

# many -to- count visualization
sns.countplot(x="gender", data=data, palette="muted"); 
plt.show()

fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.3, wspace=0.2)

sns.countplot(x="NationalITy", data=data, palette="muted");
plt.show()

fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.3, wspace=0.2)

sns.countplot(x="PlaceofBirth", data=data, palette="muted");
plt.show()

fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.3, wspace=0.5)

sns.countplot(x="StageID", data=data,  palette="muted");
plt.show()

fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.3, wspace=0.2)

sns.countplot(x="Topic", data=data,palette="bright"); 
plt.show()
```

```python
# count with ccombining more than one classes  ##
import pandas as pd
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('xAPI-Edu-Data.csv')

# defining the category variables
cat_Vars = ['gender', 'NationalITy', 'PlaceofBirth', 'StageID','Relation', 'GradeID', 'Topic', 'Semester', 'StudentAbsenceDays']

# fig specification
fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.3, wspace=0.2)

# many -to- count visualization
sns.countplot(x="Topic", data=data, hue='NationalITy',palette="bright", label=""); 
plt.show()
```

```python
# count with ccombining more than one classes  ##
import pandas as pd
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('xAPI-Edu-Data.csv')

# defining the category variables
cat_Vars = ['gender', 'NationalITy', 'PlaceofBirth', 'StageID','Relation', 'GradeID', 'Topic', 'Semester', 'StudentAbsenceDays','raisedhands']

# fig specification
fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.3, wspace=0.2)

# three classes visualization 
sns.barplot(x="Class", y="raisedhands" , hue='NationalITy', data=data,palette="bright")
plt.show()
```
